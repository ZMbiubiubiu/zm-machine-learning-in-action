# 梯度下降法

$$
w = w - \eta \frac{\partial cost}{\partial w}
$$

$$
 \frac{\partial cost}{\partial w} = \frac{1}{N} \sum_{n=1}^{N}2x_n(wx_n - y_n)
$$


```python
x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]

w = 1.0

def forward(x):
    return x * w

def cost(xs,ys):
    cost = 0
    for x,y in zip(xs,ys):
        y_pred = forward(x)
        cost += (y_pred - y) ** 2
    return cost / len(xs)

def gradient(xs,ys):
    grad = 0
    for x,y in zip(xs,ys):
        grad += 2 * x * (x * w - y)
    return grad / len(xs)

print("predict (before training)",4,forward(4))

epoch_list = []
cost_list = []

for epoch in range(100):
    cost_val = cost(x_data,y_data)
    grad_val = gradient(x_data,y_data)
    epoch_list.append(epoch)
    cost_list.append(cost_val)
    
    w -= 0.01 * grad_val
    print("epoch:",epoch,"w:",round(w,2),"loss:",round(cost_val,2)) # 保存两位小数

print("predict (after training)",4,forward(4)) 

# 绘制每个epoch的cost图

import matplotlib.pyplot as plt

plt.plot(epoch_list,cost_list)
plt.ylabel('cost')
plt.xlabel('epoch')
plt.show()
```

# 随机梯度下降法

$$
w = w - \eta \frac{\partial loss}{\partial w}
$$

$$
 \frac{\partial loss}{\partial w} = 2x_n(wx_n - y_n)
$$


```python
x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]

w = 1.0

def forward(x):
    return x * w

def loss(xs,ys):
    return (forward(xs) - ys) ** 2 

def stochastic_gradient(x,y):
    # 随机梯度下降法
    return 2 * x * (x * w - y)

print("predict (before training)",4,forward(4))

for epoch in range(100):
    for x,y in zip(x_data,y_data):
        grad_val = stochastic_gradient(x,y)
        w -= 0.01 * grad_val
        print("\tgrad:",grad_val)
    print("epoch:",epoch,"w:",round(w,2),"\n") # 保存两位小数

print("predict (after training)\n",4,forward(4)) 
```
